# Data Preprocessing

```python
import pandas as pd # data data processing
import numpy as np #linear algebra

#import data
data=pd.read_csv("./database/star_classification.csv")
```

```python
data
```

```python
#check info
data.info()
```

```python
# handle none numaric data
# GALAXY= 0, STAR=1, QSO=2.
data["class"]=[0 if i == "GALAXY" else 1 if i == "STAR" else 2 for i in data["class"]]
```

```python
#visiualisation on row data
import matplotlib.pyplot as plt
class_counts= data["class"].value_counts()

#creating a chart
plt.figure(figsize= (4,2.5))
plt.bar(class_counts.index, class_counts.values, color='skyblue')
plt.title('Distribution of Stellar Classes in Raw Data', fontsize=13)
plt.xlabel('Stellar Class', fontsize=11)
plt.ylabel('Count', fontsize=11)
plt.xticks(ticks=[0, 1, 2], labels=['GALAXY', 'STAR', 'QSO'], fontsize=9)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
data.describe()
```

```python
from sklearn.neighbors import LocalOutlierFactor
clf = LocalOutlierFactor()
y_pred = clf.fit_predict(data) 
x_score = clf.negative_outlier_factor_
outlier_score = pd.DataFrame()
outlier_score["score"] = x_score

#threshold
threshold2 = -1.5                                            
filtre2 = outlier_score["score"] < threshold2
outlier_index = outlier_score[filtre2].index.tolist()
```

```python
print(len(outlier_index))
```

```python
data.drop(outlier_index, inplace=True)
```

```python
#feature selection 
import seaborn as sns
f,ax = plt.subplots(figsize=(12,8))
sns.heatmap(data.corr(), cmap="PuBu", annot=True, linewidths=0.5, fmt= '.2f',ax=ax)
plt.show()
```

```python
corr = data.corr()
corr["class"].sort_values()
```

```python
print(data.columns)
```

```python
data = data.drop(['obj_ID','alpha','delta','run_ID','rerun_ID','cam_col','field_ID','fiber_ID'], axis = 1)
```

```python
print(data.columns)
```

```python
from sklearn.model_selection import train_test_split
x = data.drop(['class'], axis = 1)
y = data.loc[:,'class'].values
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.4, random_state = 42)
data
```

```python
#reducing training size for effeciency
x_train= x_train[:500]
y_train= y_train[:500]
```

```python
print(x_train.shape)
```

```python
print(y_train.shape)
```

```python
#Dealing with Imbalanced Data
from imblearn.over_sampling import SMOTE
from collections import Counter

x = data.drop(['class'], axis = 1)
y = data.loc[:,'class'].values


sm = SMOTE(random_state=42)
print('Original dataset shape %s' % Counter(y))
x, y = sm.fit_resample(x, y)
print('Resampled dataset shape %s' % Counter(y))
```

```python
sns.countplot(x=y, palette='Set3')
plt.title("Class Distribution", fontsize=10)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x)
x = scaler.transform(x)
```

# Data Training

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier

logreg = LogisticRegression()
svm = SVC(probability=True)
mlp = MLPClassifier(random_state=42, max_iter=500)
```

```python
# Fitting Logistic Regression model
logreg.fit(x_train, y_train)
```

```python
#Fitting Support Vector Machine model
svm.fit(x_train, y_train)
```

```python
#Fitting Multi-Layer Perceptron model
mlp.fit(x_train, y_train)
```

```python
#Support Vector Machine training model
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV 

svm_param_grid= {
    "C": [0.1, 1, 10],
    "kernel": ["linear", "rbf"],
    "gamma": ["scale", "auto"]
}
svm= SVC(probability=True)
svm_grid_search= GridSearchCV(svm, svm_param_grid, cv=3, scoring="f1", verbose=1)
svm_grid_search.fit(x_train, y_train)
```

```python
#Logistic Regression training model
from sklearn.linear_model import LogisticRegression

lr_param_grid= {
    "C": [0.1, 1, 10],
    "penalty": ["l1", "l2", "elasticnet", "none"],
    "solver": ["liblinear", "saga"]
}
logreg= LogisticRegression(max_iter=500)
lr_grid_search= GridSearchCV(logreg, lr_param_grid, cv=5, scoring="f1", verbose=1)
lr_grid_search.fit(x_train, y_train)
```

```python
#Multi-Layer Perceptron training model
mlp_params= {
    "hidden_layer_sizes": [(50,), (100,), (100,50)],
    "activation": ["relu", "tanh"],
    "solver": ["adam", "sgd"],
    "alpha": [0.0001, 0.001, 0.01]
}
mlp_grid_search= GridSearchCV(mlp, mlp_params, scoring="accuracy", cv=3)
mlp_grid_search.fit(x_train, y_train)
```

```python
best_svm= svm_grid_search.best_estimator_
best_svm
```

```python
best_lr= lr_grid_search.best_estimator_
best_lr
```

```python
best_mlp=mlp_grid_search.best_estimator_
best_mlp
```

# Model evaluation

```python
y_pred_logreg=logreg.predict(x_test)
print(f"Logistic Regression Predictions: {y_pred_logreg[:10]}")
```

```python
y_pred_svm = svm.predict(x_test)
print(f"SVM Predictions: {y_pred_svm[:10]}")
```

```python
y_pred_mlp= mlp.predict(x_test)
print(f"MLP Prediction: {y_pred_mlp[:10]}")
```

```python
models= {"MLP": best_mlp, "SVM": best_svm, "Logistic Regression": best_lr}
for model_name, model in models.items():
    print(f"Evaluating {model_name}...")
    y_pred= model.predict(x_test)
    y_pred_proba= model.predict_proba(x_test)[:,1]
```

```python
print(y_pred_proba.shape)
```

```python
y_pred_proba_svm=svm.predict_proba(x_test)
print(y_pred_proba_svm.shape)
```

```python
y_pred_proba_logreg=logreg.predict_proba(x_test)
print(y_pred_proba_logreg.shape)
```

```python
y_pred_proba_mlp= mlp.predict_proba(x_test)
print(y_pred_proba_mlp.shape)
```

```python
#Logistic Regression metrics
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred_logreg, average="weighted")
recall = recall_score(y_test, y_pred_logreg, average="weighted")
f1 = f1_score(y_test, y_pred_logreg, average="weighted")
auc_roc = roc_auc_score(y_test, y_pred_proba_logreg, multi_class="ovr")

model_name = "Logistic Regression"
print(f"\n{model_name} Metrics:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
print(f"AUC-ROC: {auc_roc:.4f}")
```

```python
#Support Vector Machine metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred_svm, average="weighted")
recall = recall_score(y_test, y_pred_svm, average="weighted")
f1 = f1_score(y_test, y_pred_svm, average="weighted")
auc_roc = roc_auc_score(y_test, y_pred_proba_svm, multi_class="ovr")

model_name = "Support Vector Machine"
print(f"\n{model_name} Metrics:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
print(f"AUC-ROC: {auc_roc:.4f}")
```

```python
#Multi-Layer Perceptron metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred_mlp, average="weighted")
recall = recall_score(y_test, y_pred_mlp, average="weighted")
f1 = f1_score(y_test, y_pred_mlp, average="weighted")
auc_roc = roc_auc_score(y_test, y_pred_proba_mlp, multi_class="ovr")

model_name = "Multi-Layer Perceptron"
print(f"\n{model_name} Metrics:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
print(f"AUC-ROC: {auc_roc:.4f}")
```

```python
#Confusion Matrix for logistic regression
conf_matrix= confusion_matrix(y_test, y_pred_logreg)
disp=ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=np.unique(y))
disp.plot(cmap=plt.cm.Blues)
plt.title(f"Confusion Matrix for {logreg}")
plt.show
```

```python
#Confusion Matrix for Multi-Layer Perceptron
conf_matrix= confusion_matrix(y_test, y_pred_svm)
disp=ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=np.unique(y))
disp.plot(cmap=plt.cm.Blues)
plt.title(f"Confusion Matrix for {svm}")
plt.show
```

```python
#Confusion Matrix for Multi-Layer Perceptron
conf_matrix= confusion_matrix(y_test, y_pred_mlp)
disp=ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=np.unique(y))
disp.plot(cmap=plt.cm.Blues)
plt.title(f"Confusion Matrix for {mlp}")
plt.show
```
